```python
import seaborn as sns
import matplotlib.pyplot as plt
#load example dataset
flights=sns.load_dataset("flights")
print(flights)
#plt.legend(loc='best')
sns.violinplot(x="year",y="passengers",data=flights)
plt.title("year Vs passengers")
plt.xlabel("year")
plt.ylabel("passengers")
#plt.legenda(loc='best')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load example dataset
exercise=sns.load_dataset("exercise")
print(exercise)
plt.legend(loc='best')
sns.violinplot(x="pulse",y="time",data=exercise,label="exercise")
plt.title("measurement of pulse rate with respect to time")
plt.xlabel("pulse")
plt.ylabel("time")
#plt.legenda(loc='best')
plt.show()
```
